#1. Write a Python program to reverse words of sentence in the same orders

```python
def reverse_words(sentence):
    words = sentence.split()
    reversed_words = []
    
    for word in words:
        reversed_word = word[::-1]
        reversed_words.append(reversed_word)
    
    reversed_sentence = ' '.join(reversed_words)
    return reversed_sentence
```

```python

# Example usage
sentence = "Python programming is fun"
print(reverse_words(sentence))
```

#2 Write a Python program to reverse words of sentence in the same order 

```python
def powerset(s):
    result = [[]]
    
    for element in s:
        for subset in result[:]:
            new_subset = subset + [element]
            result.append(new_subset)
    
    return result
```

```python
# Example usage
my_set = [1, 2, 3]
print(powerset(my_set))
```

#3 Write a Python code to to pickup a list from a string dynamically

```python
import ast

def extract_list(string):
    try:
        extracted_list = ast.literal_eval(string)
        if isinstance(extracted_list, list):
            return extracted_list
        else:
            return "No list found in the string."
    except:
        return "Invalid input."

# Example usage
string = "[1, 2, 3, 4]"
print(extract_list(string))
```

#4 Write a Python fuction which takes "Text, chunk size, overlap size" as argument to divide the given text into chunks of given size with given overlap size

```python
def chunk_text(text, chunk_size, overlap_size):
    chunks = []
    start = 0
    length = len(text)
    
    while start < length:
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start = start + chunk_size - overlap_size
        
    return chunks

# Example usage
text = "Hello, how are you doing today?"
chunk_size = 10
overlap_size = 3
print(chunk_text(text, chunk_size, overlap_size))
```

 Write a Python function to accept an input string and perform the following operations on it:
1. Coaunt the total number of words in the sentence
2. Find the longest word in the sentence
3. Return a new sentence in which: 
    a. each word starts with an uppercase letter
    b. All extra spaces (leading, trailing, or multiple spaces between words) are removed.

Final output should be a Tuple with the above three elements.

```python
def process_sentence(sentence):
    # Remove extra spaces and split into words
    words = sentence.strip().split()
    
    # Count words
    word_count = len(words)
    
    # Find the longest word
    longest_word = ""
    for word in words:
        if len(word) > len(longest_word):
            longest_word = word
    
    # Create new sentence with each word capitalized
    capitalized_words = []
    for word in words:
        capitalized_words.append(word.capitalize())
    new_sentence = " ".join(capitalized_words)
    
    return (word_count, longest_word, new_sentence)

# Example usage
sentence = "   hello   world   from    python  "
print(process_sentence(sentence))
```
